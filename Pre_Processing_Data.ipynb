# **Rename Columns**

```python
df.rename(columns={'Order Id':'Order_ID'}, inplace=True)
df.rename(columns={'Type':'Payment_Method'}, inplace=True)
df.rename(columns={'Days for shipping (real)':'Real_Days_Shipping'}, inplace=True)
df.rename(columns={'Days for shipment (scheduled)':'Scheduled_Days_Shipping'}, inplace=True)
df.rename(columns={'Customer Id':'Customer_ID'}, inplace=True)
df.rename(columns={'Customer City':'Customer_City'}, inplace=True)
df.rename(columns={'Customer Country':'Customer_Country'}, inplace=True)
df.rename(columns={'Customer Fname':'Customer_Fname'}, inplace=True)
df.rename(columns={'Customer Lname':'Customer_Lname'}, inplace=True)
df.rename(columns={'Customer Segment':'Customer_Segment'}, inplace=True)
df.rename(columns={'Customer State':'Customer_State'}, inplace=True)
df.rename(columns={'Customer Street':'Customer_Street'}, inplace=True)
df.rename(columns={'Department Name':'Department_Name'}, inplace=True)
df.rename(columns={'Market':'Order_Market'}, inplace=True)
df.rename(columns={'Order Region':'Order_Region'}, inplace=True)
df.rename(columns={'Order State':'Order_State'}, inplace=True)
df.rename(columns={'Order City':'Order_City'}, inplace=True)
df.rename(columns={'Order Country':'Order_Country'}, inplace=True)
df.rename(columns={'order date (DateOrders)':'Order_Date'}, inplace=True)
df.rename(columns={'Order Item Cardprod Id':'Item_ID'}, inplace=True)
df.rename(columns={'Category Name':'Category_Name'}, inplace=True)
df.rename(columns={'Product Name':'Item_Name'}, inplace=True)
df.rename(columns={'Order Item Product Price':'Item_Price'}, inplace=True)
df.rename(columns={'Order Item Discount':'Item_Discount'}, inplace=True)
df.rename(columns={'Order Item Quantity':'Quantity'}, inplace=True)
df.rename(columns={'Sales':'Estimate_Value'}, inplace=True)
df.rename(columns={'Order Item Total':'Real_Value'}, inplace=True)
df.rename(columns={'Order Profit Per Order':'Order_Profit'}, inplace=True)
df.rename(columns={'Order Status':'Order_Status'}, inplace=True)
df.rename(columns={'shipping date (DateOrders)':'Shipping_Date'}, inplace=True)
df.rename(columns={'Shipping Mode':'Shipping_Mode'}, inplace=True)
```

# **Change Datetimes type**

```python
df['Shipping_Date'] = pd.to_datetime(df['Shipping_Date'], format='%m/%d/%Y %H:%M')
df['Shipping_Date'] = df['Shipping_Date'].dt.strftime('%d-%m-%Y')
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%m/%d/%Y %H:%M')
df['Order_Date'] = df['Order_Date'].dt.strftime('%d-%m-%Y')
```

```python
df['Shipment_ID'] = df.apply(lambda row: row['Order_ID'], axis=1)
df.head()
```

# **Add Shipment ID**

```python
cols = ['Order_ID',
  'Shipment_ID',
 'Payment_Method',
 'Real_Days_Shipping',
 'Scheduled_Days_Shipping',
 'Customer_ID',
 'Customer_City',
 'Customer_Country',
 'Customer_Fname',
 'Customer_Lname',
 'Customer_Segment',
 'Customer_State',
 'Customer_Street',
 'Department_Name',
 'Order_Market',
 'Order_Region',
 'Order_State',
 'Order_City',
 'Order_Country',
 'Order_Date',
 'Item_ID',
 'Category_Name',
 'Item_Name',
 'Item_Price',
 'Item_Discount',
 'Quantity',
 'Estimate_Value',
 'Real_Value',
 'Order_Profit',
 'Order_Status',
 'Shipping_Date',
 'Shipping_Mode']

df = df[cols]
```

# **Modify Order ID**

```python
df['Order_ID'] = df.index.to_list()
df.head()
```
